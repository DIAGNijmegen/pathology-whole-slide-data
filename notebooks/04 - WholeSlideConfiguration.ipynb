```python
from wholeslidedata.configuration.config import WholeSlideDataConfiguration
from matplotlib import pyplot as plt
USER_CONFIG = './configs/user_config.yml'
MODES = ('training', )
```

```python
build = WholeSlideDataConfiguration.build(user_config=USER_CONFIG, modes=MODES)

training_dataset = build['wholeslidedata']['training']['dataset']
training_batch_sampler = build['wholeslidedata']['training']['batch_sampler']
training_batch_shape = build['wholeslidedata']['training']['batch_shape']

spacing = training_batch_shape.spacing
shape = training_batch_shape.shape
```

```python
training_batch = training_batch_sampler.batch([training_dataset.sample_references['lymphocytes'][0],
                                               training_dataset.sample_references['lymphocytes'][1]])
x_batch, y_batch = training_batch
```
